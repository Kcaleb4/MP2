```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
stockData = pd.read_csv("Stock_Price_Bank.csv")
stockData.head()
```

```python
stockData.shape
```

```python
stockData['Date'] = pd.to_datetime(stockData['Date'])
```

```python
fig, axs = plt.subplots(3, figsize=(10, 17))

axs[0].scatter(stockData["High"], stockData["Volume"])
axs[0].set_xlabel("High")
axs[0].set_ylabel("Volume")
axs[0].set_title("Scatter Plot of High vs Volume")

axs[1].scatter(stockData["Date"], stockData["Open"])
axs[1].set_xlabel("Date")
axs[1].set_ylabel("Open")
axs[1].set_title("Scatter Plot of Date vs Open")

axs[2].scatter(stockData["High"], stockData["Low"])
axs[2].set_xlabel("High")
axs[2].set_ylabel("Low")
axs[2].set_title("scatter Plot of High vs Low")

plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Assuming stockData is your DataFrame with columns 'Date' and 'Price'
growthByYear = stockData.groupby(stockData['Date'].dt.year)['Open'].pct_change().groupby(stockData['Date'].dt.year).mean() * 100
growthByYear
```

```python
plt.rcParams["text.color"] = "black"
plt.rcParams["figure.facecolor"] = "#DEDCA9"
plt.figure(figsize=(8, 8))
plt.rcParams["font.size"] = 15
plt.title("Growth % Per Year")
plt.pie(growthByYear, labels = growthByYear.index, autopct = '%1.0f%%', colors = ["#23DB39", "#23DB77","#4EDB18","#B6DB63","#7FE08A","#A4E0D4","#32A855","#B2AF85","#52DB80","#49A85B","#52BD6D"])
plt.show
```

```python
plt.hist(stockData["Close"], color = "green")
plt.xlabel("Close")
plt.ylabel("frequency")
plt.title("Close Price Distribution")
plt.show()
```

```python
plt.hist(stockData["Low"], color = "green")
plt.xlabel("Lows")
plt.ylabel("frequency")
plt.title("Daily Low Distribution")
plt.show()
```
